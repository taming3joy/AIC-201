```python
#Load iris dataset
from sklearn.datasets import load_iris
df=load_iris(as_frame=True).frame
print (df.head())
```

```python
#Separate data and target into X and y
X=df.iloc[:,:-1] #All rows, every column except last
y=df.iloc[:,-1] #All rows, last column only
print (X.head())
print (y.head())
```

```python
#Split dataset into Test and Train
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
#Accuracy testing with Cross Validation
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,X_train,y_train,cv=5)
print (f"Mean accuracy: {scores.mean():.2f}")
print (f"Standard deviation: {scores.std():.2f}")
```

```python
#Visualization of the decision tree
from sklearn.tree import plot_tree
model.fit(X_train,y_train)
plot_tree(model)
```

```python
#Parameters tuning
from sklearn.model_selection import GridSearchCV
parameters_grid={
    "criterion": ["gini", "entropy", "log_loss"],
    "max_depth": [None, 3, 5, 8, 12, 16, 24],
    "min_samples_split": [2, 5, 10, 20, 50],
    "min_samples_leaf": [1, 2, 5, 10, 20],
    "max_features": [None, "sqrt", "log2"],
    "class_weight": [None, "balanced"],
}

grid=GridSearchCV(model,parameters_grid,cv=5)
grid.fit(X_train,y_train)
```

```python
print (grid.best_params_)
```

```python
#Accuracy testing with Cross Validation on tuned model
tuned_model = grid.best_estimator_
new_scores=cross_val_score(tuned_model,X_train,y_train,cv=5)
print (f"Mean accuracy: {new_scores.mean():.2f}")
print (f"Standard deviation: {new_scores.std():.2f}")
```

```python
#Change model to kNN
from sklearn.neighbors import KNeighborsClassifier
kNN=KNeighborsClassifier(n_neighbors=4)
kNN_scores = cross_val_score(kNN,X_train,y_train,cv=5)
print (f"Mean accuracy: {kNN_scores.mean():.2f}")
print (f"Standard deviation: {kNN_scores.std():.2f}")
```
